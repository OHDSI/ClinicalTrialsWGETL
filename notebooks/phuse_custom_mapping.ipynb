## PHUSE_AE_AESEV

```
# CRF p.121
%%sql

SELECT 'Severity/Intensity'
       || '|'
       || aesev AS source_code,
       'Severity/Intensity'
       || '|'
       || aesev AS source_description,
       Count(*) AS cnt
FROM   src.ae
WHERE  aesev IS NOT NULL
GROUP  BY 'Severity/Intensity'
          || '|'
          || aesev
ORDER  BY source_code; 
```

## PHUSE_CM_CMTRT

```
%%sql

SELECT cmtrt 
       || '|' 
       || cmdose 
       || '|' 
       || Coalesce(cmdosu, '') 
       || '|' 
       || Coalesce(cmroute, '') AS source_code, 
       Coalesce(NULLIF(cmdecod, 'UNCODED'), cmtrt) 
       || '|' 
       || 'dose: ' 
       || cmdose 
       || '|' 
       || Coalesce(cmdosu, '') 
       || '|' 
       || Coalesce(cmroute, '') AS source_description, 
       Count(*)                 AS cnt 
FROM   src.cm 
WHERE  cmtrt IS NOT NULL 
GROUP  BY cmtrt 
          || '|' 
          || cmdose 
          || '|' 
          || Coalesce(cmdosu, '') 
          || '|' 
          || Coalesce(cmroute, ''), 
          Coalesce(NULLIF(cmdecod, 'UNCODED'), cmtrt)  
          || '|' 
          || 'dose: ' 
          || cmdose 
          || '|' 
          || Coalesce(cmdosu, '') 
          || '|' 
          || Coalesce(cmroute, '') 
ORDER  BY cnt DESC; 
```
